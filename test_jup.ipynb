```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield
        a, b = b, a + b

# Использование
for num in fibonacci(5):
    print(num)  # 0,1,1,2,3
```

```python
import argparse
import csv
from collections import defaultdict

students_dict = defaultdict(list)

with open('resourses/math.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for index, row in enumerate(reader):
        students_dict[row['student']].append(int(row['coffee_spent'].replace(' ', '0')))

print(students_dict)
```

```python
for i in students_dict:
    print(i)
```
